````python
def get_SangGwan(api_key, start, end, year, svc_induty_cd=''):
    url = f"http://openapi.seoul.go.kr:8088/{api_key}/xml/VwsmTrdhlStorQq/{start}/{end}/{year}/{svc_induty_cd}"
    request = urllib.request.Request(url)
    response = urllib.request.urlopen(request)
    cc = xmltodict.parse(response.read().decode('utf-8')) # return collections.OrderedDict
    dd = json.loads(json.dumps(cc)) # return dict
    return dd['VwsmTrdhlStorQq']

def get_all_SangGwan(api_key, year, max_cnt=None, svc_induty_cd=''):
    if max_cnt == None:
        total_cnt = int(get_SangGwan(api_key, 1, 1, year).get('list_total_count'))
        print(f'총 {total_cnt} 존재')
    else:
        total_cnt = max_cnt
        print(f'총 {total_cnt}건에 대해 api 호출')
    ls = []
    for start in range(1, total_cnt, 1000):
        get_item = get_SangGwan(api_key, start, start+999, year, svc_induty_cd)
        ls.extend(get_item['row'])
        print(f'{start} ~ {start+999} 추출 완료')
    print("끝!!!")
    return pd.DataFrame(ls)
````

````python
# 2013년부터 있음 but 개수가 적은듯
url = "http://openapi.seoul.go.kr:8088/"+my_key+"/xml/VwsmTrdarSelngQq/1/5/2017/A"
request = urllib.request.Request(url)
response = urllib.request.urlopen(request)
cc = xmltodict.parse(response.read().decode('utf-8')) # return collections.OrderedDict
dd = json.loads(json.dumps(cc)) # return dict
dd
````

````python
def get_Sales(api_key, start, end, year):
    url = f"http://openapi.seoul.go.kr:8088/{api_key}/xml/VwsmTrdarSelngQq/{start}/{end}/{year}"
    request = urllib.request.Request(url)
    response = urllib.request.urlopen(request)
    cc = xmltodict.parse(response.read().decode('utf-8')) # return collections.OrderedDict
    dd = json.loads(json.dumps(cc)) # return dict
    return dd['VwsmTrdarSelngQq']

def get_all_Sales(api_key, year, max_cnt=None):
    if max_cnt == None:
        total_cnt = int(get_Sales(api_key, 1, 1, year).get('list_total_count'))
        print(f'총 {total_cnt} 존재')
    else:
        total_cnt = max_cnt
        print(f'총 {total_cnt}건에 대해 api 호출')
    ls = []
    for start in range(1, total_cnt, 1000):
        get_item = get_Sales(api_key, start, start+999, year)
        ls.extend(get_item['row'])
        print(f'{start} ~ {start+999} 추출 완료')
    return pd.DataFrame(ls)
````

````python
df_sales_ls = []
for year in range(2014, 2022):
    df_sales = get_all_Sales(my_key, year)
    df_sales_ls.append(df_sales)
````

````python
# THSMON_SELNG_AMT 분기당_매출_금액
# THSMON_SELNG_CO 분기당_매출_건수
def get_industry_sales(df_sales, industry):
    df_industry = df_sales[df_sales['SVC_INDUTY_CD_NM']==industry][['THSMON_SELNG_AMT','THSMON_SELNG_CO']]
    df_industry.rename(columns={"THSMON_SELNG_AMT": industry + "_분기당_매출_금액", 
                                "THSMON_SELNG_CO": industry + "_분기당_매출_건수"},
                       inplace=True)
    return df_industry

def get_all_industry_sales(df_sales, need_industry_ls):
    df = pd.DataFrame(index=df_sales.index.unique())
    for industry in need_industry_ls:
        df_industry = get_industry_sales(df_sales, industry)
        df = df.merge(df_industry, how='left', left_index=True, right_index=True)
        
    return df
        
````

````python
cnt_dic = {}
def count_store(row):
    dong = row['소재지전체주소'].split()[2]
    year = int(row['인허가일자'][:4])
    q = (int(row['인허가일자'][4:6]) - 1) // 3 + 1
    
    if row['영업상태명'] == '폐업':
#         print('시작일 : ',row['인허가일자'], year, q, end='\t')
        end_year = int(row['폐업일자'][:4])
        end_q = (int(row['폐업일자'][4:6]) - 1) // 3 + 1
#         print('종료일 : ',row['폐업일자'], end_year, end_q)
    else:
        end_year = 2021
        end_q = 2
#         print('시작일 : ',row['인허가일자'], year, q)
    
    if year < 2014:
        year, q= 2014, 1
    for y in range(year, end_year+1):
        start_q = q if y == year else 1
        e_q = end_q if y == end_year else 4
            
            
        for qq in range(start_q, e_q+1):
            idx = dong+'_'+str(y)+'_'+str(qq)
            if cnt_dic.get(idx):
                cnt_dic[idx][0] += 1
            else:
                cnt_dic[idx] = [1]
            
                    
        
````

````python
for idx, row in star_bucks_df.iterrows():
    count_store(row)

cnt_dic
````

````python
star_bucks_cnt_df = pd.DataFrame(cnt_dic).transpose()
star_bucks_cnt_df.columns=['스타벅스매장수']
star_bucks_cnt_df
````
